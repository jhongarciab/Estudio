```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
# Cargar dataset (ya subido en este entorno como kc_house_data.csv)
df = pd.read_csv("kc_house_data.csv")
df.head()
```

```python
# Mostrar los tipos de datos de cada columna
df.dtypes
```

```python
# Eliminar columnas id y Unnamed:0
df.drop(["id", "Unnamed: 0"], axis=1, inplace=True, errors='ignore')

# Resumen estadístico
df.describe()
```

```python
# Contar casas por número de pisos y pasarlo a DataFrame
df["floors"].value_counts().to_frame()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="waterfront", y="price", data=df)
plt.title("Precio de casas con o sin vista al waterfront")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="sqft_above", y="price", data=df, line_kws={"color":"red"})
plt.title("Relación entre sqft_above y price")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df[["sqft_living"]]
y = df["price"]

lm = LinearRegression()
lm.fit(X, y)

# Calcular R^2
lm.score(X, y)
```

```python
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement",
            "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

X_multi = df[features]
y = df["price"]

lm_multi = LinearRegression()
lm_multi.fit(X_multi, y)

# Calcular R^2
lm_multi.score(X_multi, y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

pipeline = Pipeline([
    ("scale", StandardScaler()),
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("model", LinearRegression())
])

pipeline.fit(X_multi, y)

# Calcular R^2
pipeline.score(X_multi, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge

X_train, X_test, y_train, y_test = train_test_split(X_multi, y, test_size=0.2, random_state=1)

ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_train, y_train)

# Calcular R^2 en test
ridge_model.score(X_test, y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

poly2 = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly2.fit_transform(X_train)
X_test_poly = poly2.transform(X_test)

ridge_poly = Ridge(alpha=0.1)
ridge_poly.fit(X_train_poly, y_train)

# R^2 en test
ridge_poly.score(X_test_poly, y_test)
```
